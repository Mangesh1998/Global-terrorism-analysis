## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#put variables in List format 
data.columns.to_list()
````

````python
#Rename variables that are to be consider for analysis 
data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day', 'country_txt':'Country','provstate':'state',
                         'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type'},inplace=True)
data.columns
````

````python
# Consider only those variables in dataset which are to be taken for further analysis
df = data[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed',
                      'Wounded','Target','Group','Target_type','Weapon_type']]
df.head()
````

````python
df.shape
````

````python
#where exactly missing values are?
df.isnull().sum()
````

Data shows that there are missing values in variables like state,city,latitude,longitude,killed,Wounded,Target            

 Firstly  drop missing values of variables state,city,latitude,longitude,Target

````python
#use drop.na method
print(df.shape)
df.dropna(subset=['state','city','latitude','longitude','Target'],axis=0,inplace=True)
#inplace=True modifies the data frame
df.isnull().sum()
````

Then Replace missing values with the average of that column for variables like Killed ,Wounded

````python
mean=df['Killed'].mean()
mean=df['Wounded'].mean()

#Replacing NaN values with mean of the data(updating as well)
df['Killed']=df['Killed'].replace(np.nan,mean)

df['Wounded']=df['Wounded'].replace(np.nan,mean)
df.isnull().sum()
````

````python
df.isnull().sum()
````

````python
df.shape
````

### What all manipulations have you done and insights you found?

Initially for better understanding of terminologies I rename variables that are to be consider for analysis .Then consider only those variables in dataset which are to be taken for further analysis.Thus I curved out dataframe of shape (181691, 16).  Data shows that there are missing values in variables like state,city,latitude,longitude,killed,Wounded,Target so I Firstly drop missing values of variables state,city,latitude,longitude,Target and then replace missing values with the average of that column for variables like Killed ,Wounded.Now there are no missing values in our dataset.And we are ready to analyse data with interesting  plot vizualizations .Now there are no NaN values and we obtained dataframe of shape (175708, 16)
